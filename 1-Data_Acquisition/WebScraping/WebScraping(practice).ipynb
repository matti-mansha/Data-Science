```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
```

```python
def getdata(url):
    response = requests.request("GET", url) 
    html_data = BeautifulSoup(response.text, 'html.parser')
    return html_data
```

```python
def getnextpage(soup):
    page = soup.find('ul', {'class': "m-list m-list--centered"})
    try: 
        if page.find('a', {'class': 'c-button c-button--hollow'}):
            url = "https://www.politifact.com/factchecks/list/" + str(page.find_all('a', {'class': 'c-button c-button--hollow'})[1].attrs['href'])
            return url
    except IndexError:
        if page.find('a', {'class': 'c-button c-button--hollow'}):
            url = "https://www.politifact.com/factchecks/list/" + str(page.find('a', {'class': 'c-button c-button--hollow'}).attrs['href'])
            return url
    else:
        return
```

```python
url = "https://www.politifact.com/factchecks/list/?page=1"
dic_list = []

while url is not None:
    print(url)
    soup = getdata(url)
    boxes = soup.find_all('li', {'class':'o-listicle__item'})
    for box in boxes:
        try:
            dic = {}
            dic['Title'] = box.find_all('a')[0].attrs['title']

            dic['StartDate'] = re.findall('[A-Za-z]+ [0-9]+, [0-9]+', box.find_all('div', {'class':'m-statement__desc'})[0].text.strip())[0]

            dic['Link'] = "https://www.politifact.com" + box.find_all('a')[1].attrs['href']

            dic['Post'] = box.find_all('a')[1].text.strip().replace('“', '').replace('”', '')
            
            dic['Status'] = box.find('div', {'class':'m-statement__meter'}).img.attrs['alt']
            

            dic_list.append(dic)
        except AttributeError:
            pass
    url = getnextpage(soup)
```

```python
df = pd.DataFrame(dic_list)
```
