#### Importing important Data Science libraries

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

#### Method to get HTML script

```python
def getdata(url):
    request = Request(url)
    page = urlopen(request)
    html_data = BeautifulSoup(page, 'html.parser')
    return html_data
```

#### Get next page link

```python
def getnextpage(soup):
    page = soup.find('nav', {'class':'pagination'})
    if page.find('span', {'class': 'next'}):
        url = 'https://www.passiton.com/' + str(page.find('span', {'class': 'next'}).find('a')['href'])
        return url
    else:
        return
```

```python
url = "https://www.passiton.com//inspirational-quotes"
dic_list = []

while url is not None:
    print(url)
    soup = getdata(url)
    boxes = soup.find_all('div', {'class':'col-6 col-lg-4 text-center margin-30px-bottom sm-margin-30px-top'})
    for box in boxes:
        try:
            dic = {}
            dic['Category'] = box.h5.text
            dic['Quote'] = box.img.attrs['alt'].split('#')[0]
            dic['Image-link'] = box.img.attrs['src']
            dic['Quote-url'] = box.a.attrs['href']
            dic_list.append(dic)
        except AttributeError:
            pass
    url = getnextpage(soup)
```

#### Dataframe maker

```python
df = pd.DataFrame(dic_list)
```
